Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and open-source Python web framework used to build web applications and web APIs.

It follows a micro-framework approach (only essential features are provided; you can add others as needed).

✅ Advantages of Flask

Simple and easy to learn

Lightweight and fast

Very flexible and extensible

Built-in development server and debugger

Good support for REST APIs

Easy integration with databases and other libraries

Python code

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello():
    return "Hello World!!"

if __name__ == "__main__":
    app.run()
```

Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask is the process of mapping a URL to a Python function.

It is done using:

```python
@app.route("/path")
```

Why app routes are used

To decide which function should run for a given URL

To create different pages (endpoints) in a web application

To handle user requests such as /login, /about, etc.

Example:

```python
@app.route("/home")
def home():
    return "This is home page"
```

**Q5. What function is used in Flask for URL building?

Write a Python code to demonstrate url_for()**

The function used for URL building in Flask is:

✅ url_for()

It is used to generate URLs dynamically using the function name instead of hard-coding paths.

Example code

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route("/")
def home():
    return "Home Page"

@app.route("/about")
def about():
    return "About Page"

@app.route("/showurl")
def show_url():
    return url_for('about')

if __name__ == "__main__":
    app.run()
```

Explanation

url_for('about') automatically builds the URL for the function about.

Output of /showurl will be:

```python
/about
```
